```python
import pandas as pd
```

## Limpeza de dados 2020

```python
dados_2020 = pd.read_csv(r"C:\Users\gabri\OneDrive\Documentos\2 periodo\intro. cd\projetosiconf\dados\finbraRREO2020.1.csv", sep = ";", skiprows=5, encoding="latin1")
```

```python
dados_2020.head()
```

```python
dados_2020 =dados_2020.drop(columns=['Cod.IBGE', 'Coluna'])
dados_2020
```

```python
dados_2020.insert(0, 'Ano', 2020)
dados_2020
```

```python
dados_2020['Valor'] = dados_2020['Valor'].str.replace('.', '', regex=False) 
dados_2020['Valor'] = dados_2020['Valor'].str.replace(',', '.', regex=False)

dados_2020['Valor'] = pd.to_numeric(dados_2020['Valor'], errors='coerce')

dados_2020['Valor'] = dados_2020['Valor'].apply(lambda x: f' {x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'))
```
